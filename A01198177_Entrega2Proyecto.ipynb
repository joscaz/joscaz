# Descripción problemática


Según la prueba PISA del 2012, el 55% de los adolescentes mexicanos no alcanza el nivel básico para el área de las matemáticas. (REFERENCIA). Con esto dicho, quiero enfocar mi proyecto a mejorar los conocimientos y habilidades en este sector ya que las matemáticas son muy importantes en nuestro día a día y no puede ser posible que los jóvenes de nuestro país no tengan las competencias básicas de esta área. Las matemáticas nos ayuda al pensamiento lógico, a tomar decisiones, a estar listos para pensar, a ser críticos, etc. 

Para que se pueda tener un mejor país, debemos mejorar en muchas áreas, pero en este caso me enfocaré en el area de las matemáticas. 

Mi solución es crear un programa en el cual se puedan contestar diferentes problemas matemáticos con distintos grados de dificultad y con la explicación del problema para que así los alumnos puedan practicar más y mejorar sus habilidades. También se podrá incluir alguna pregunta que tenga que ver con las otras dos áreas del PISA (lectura y ciencias)

Funcionalidad del programa:
Al principio se desplegará un menú con las diferentes opciones que el usuario puede escoger. Después se le preguntará que cuál es la opción que desea seleccionar. Ya con la opción seleccionada, se accionará la opción correspondiente. Por ejemplo, si el usuario escoge la opción 4 (la cual es presentar un quiz) se le hará un pequeño examen en el cual se podrá calificar su desempeño en matemáticas. Si tiene alguna respuesta incorrecta, se le hará saber la respuesta correcta al finalizar y con una explicación breve. 

En la opción 3 (estudiar preguntas de un área en específico) el usuario podrá escoger alguna otra área en la cual quisiera reforzar conocimientos. Primero escoge el área a estudiar, después vienen ciertos temas y conceptos fundamentales del área y al final hay un pequeño examenal respecto. 

En la opción 5 (reporte de calificaciones) se tomarán en cuenta los resultados de los pequeños exámenes que el usuario hizo, y en base a eso y al criterio de calificación de PISA, se desplegará la calificación obtenida y el desempeño que el usuario tuvo a lo largo del programa. 

La opción 1 y 2 como su nombre lo indica servirá para dar de alta o actualizar alguna pregunta simplemente. 

Cuando el usuario desee interrumpir o finalizar su evaluación, podrá escoger la opción 6 para salir.

```python
preguntas = []
pregunta_upd = None
preguntas_nueva = []
promedio = 0
suma_promedio = 0
preg_correctas = 0
preg_incorrectas = 0
prom = None

veces_func1 = 0 #veces que se ejecutó la función 1
veces_func2 = 0 #veces que se ejecutó la función 2
veces_func3 = 0 #veces que se ejecutó la función 3
veces_func4 = 0 #veces que se ejecutó la función 4
veces_func5 = 0 #veces que se ejecutó la función 5

def registrar_pregunta():
  global preguntas
  global veces_func1

  print("Bienvenido a la opción para registrar preguntas, por favor sigue las instrucciones")
  id_pregunta = input("Ingresa el ID de la pregunta: ")
  enfoque = input("Ingresa el enfoque de la pregunta: Lectura, Matemáticas, Ciencias: ")
  pregunta = input("Ingresa el texto de la pregunta: ")
  respuestas = ""

  for num in range(4):
    respuesta = input("Ingresa la respuesta de la pregunta " + str(num + 1) + ": ")

    if num == 3:
      respuestas += respuesta # string de respuesta 4, no lleva coma
    else:
      respuestas += respuesta + "," # string de respuestas 1, 2, 3
 
  respuestas = respuestas.split(",") # hacer una lista desde un string, cada elemento es separado por coma
  
  respuesta1, respuesta2, respuesta3, respuesta4 = respuestas # asignación múltiple de variables de una lista (elemento por elemento)

  correcta = input("Ingresa la respuesta correcta:")

  pregunta = [id_pregunta, enfoque, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]

  confirmacion = str(input("¿Está seguro de que quiere actualizar los datos de la pregunta?(SI/NO): "))
  confirmacion = confirmacion.upper() #Cualquier letra escrita, se pasará como mayúscula

  if confirmacion == "SI":
    preguntas.append(pregunta)
    with open("preguntas.txt", "a") as archivo:
      for item in preguntas:
        archivo.write("%s\n" % item) #Método que encontre en línea para que pase la lista al archivo sin soltar error. Me indicaba que no se puede pasar una lista a un archivo.



  elif confirmacion == "NO":
    print("No se ha dado de alta la pregunta")

  else:
    print("Dato no válido")

  with open ("preguntas.txt", "r") as leer:
    renglon = leer.readlines()
    preguntas.append(renglon[-1])

  veces_func1 += 1
  print(f"Esta función se ha llamado {veces_func1} veces")
  leer_archivo("preguntas.txt")

  
def menu():
  print("Escoge alguna de las siguientes opciones y escribe el número de opción correspondiente: ")
  print("1. Alta de preguntas de prueba PISA (lectura, matemáticas, ciencias)")
  print("2. Actualizar preguntas de la prueba PISA (lectura, matemáticas, ciencias)")
  print("3. Estudiar preguntas de cierta área (lectura, matemáticas, ciencias)")
  print("4. Presentar un quiz (lectura, matemáticas, ciencias)")
  print("5. Reporte de calificaciones")
  print("6. Salir")


def main():
  menu()

  while True:

    dato = int(input("Qué opción deseas escoger?: "))

    print("Escogiste la opcion: " + str(dato))

    if dato == 1:
      registrar_pregunta()
    
    elif dato == 2:
      actualizar_pregunta()
    
    elif dato == 3:
      estudiar_area()
    
    elif dato == 4:
      presentar_quiz()
    
    elif dato == 5:
      reportar_calificaciones()

    elif dato == 6:
      salir()
      break

    else: 
      print("Opción incorrecta \n")

def actualizar_pregunta():
  global preguntas
  global pregunta_upd
  global veces_func2
  global preguntas_nueva
  
  print("Bienvenido a la opción para actualizar preguntas, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  id_cambio = input("Ingresa el ID de la pregunta a actualizar: ")
 
  for num in range(len(preguntas)):
    
    pregunta = preguntas[num]

    id_pregunta = pregunta[0]
    
    if id_cambio == id_pregunta:
      existe = True
      no_pregunta = num
      break
    else:
      existe = False
  
  if existe == True:
    print("Si existe la pregunta")
      
    id, enfoque, texto_pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = preguntas[no_pregunta]

    if enfoque == 1:
      enfoque = "Lectura"
    elif enfoque == 2:
      enfoque = "Matemáticas"
    else:
      enfoque = "Lectura"
  
    print(f'{"ID: "}{id:<5}{"Enfoque: "}{enfoque}\n{texto_pregunta}\n{respuesta1:<15}{respuesta2:<15}{respuesta3:<15}{respuesta4}\n{"Respuesta correcta: "}{correcta}')
   
    confirmacion = input("Si deseas cambiar la pregunta escribe 'si' ")

    if confirmacion.lower() == "si":
     print("Pregunta cambiada")
  
  else:
    print("No existe la pregunta, intenta con otro ID")
    actualizar_pregunta()
  
  veces_func2 += 1
  print(f"Esta función se ha llamado {veces_func2} veces")
  leer_archivo("preguntas.txt")
  grabar_archivo("lista_nueva.txt","preguntas.txt")


def estudiar_area():

  global veces_func3
  global promedio

  area_a_estudiar = input("¿Qué área quieres estudiar?: ")

  with open("preguntas.txt", "r") as archivo:
    for linea in archivo: #Este ciclo va línea por línea del archivo de texto      
      if area_a_estudiar in linea: #Si la area que ingresó el usuario que quiere estudiar se encuentra en la lista, que imprima la pregunta
        print(linea, end="\n")

  veces_func3 += 1
  print(f"Esta función se ha llamado {veces_func3} veces")

def presentar_quiz():
  global promedio
  global suma_promedio
  global veces_func4

  import time

  def countdown(t):
    
    #Declaramos las variables globales para acumuladores que usaremos al final
    global suma_promedio 
    global preg_correctas
    global preg_incorrectas
    global prom

    while t:
      mins, secs = divmod(t, 60)
      timer = '{:02d}:{:02d}'.format(mins, secs)
      print(timer, end="\r")
      time.sleep(1)
      t -= 1
      print("Estás a punto de comenzar un quiz. Tienes un límite de 20 minutos")


      with open("preguntas.txt","r") as archivo:

        #Por cada renglon en el archivo de texto
        for linea in archivo:
          linea = linea.split(",") #Que cada elemento esté dividido cuando haya una coma
          texto = (f"Esta pregunta, está relacionada al área de{(linea[1])}\n{linea[2]}\na){linea[3]:<15}b){linea[4]:<15}c){linea[5]:<15}d){linea[6]:<15}\n") #lo guardo en una variable para después borrar las comillas y que en el quiz no aparezcan y se vea limpio
          print(texto.replace("'", "")) #reemplazar las comillas simples con un string vacío 

          #Este bloquecito lo hice para cersiorarme que no hubiera espacios en blanco en los elementos. Estuve como 3 horas batallando por un error así.
          resp1 = linea[3].replace("'","").strip()
          resp2 = linea[4].replace("'","").strip()
          resp3 = linea[5].replace("'","").strip()
          resp4 = linea[6].replace("'","").strip()

          #resp1 = resp1.strip()
          #resp2 = resp2.strip()
          #resp3 = resp3.strip()
          #resp4 = resp4.strip()

          correcta = linea[7] #declaramos la posicion de la respuesta correcta en el renglon
          correcta = correcta.replace("'","")
          correcta_sola = correcta.replace("]","") #creo una variable llamada correcta_sola para desplegar en el fstring de la explicación si se saca mal.
          correcta = correcta_sola.strip() #Pongo esta linea de codigo porque por alguna razon aparecia un espacio en blanco antes del elemento correcta

          #Cambiamos el valor de la variable correcta para que así el usuario pueda ingresar la letra de la opción y no el valor completo, simulando un quiz real.
          if correcta == resp1:
            correcta = "a"
          elif correcta == resp2:
            correcta = "b"       
          elif correcta == resp3:
            correcta = "c"
          elif correcta == resp4:
            correcta = "d"

          respuesta_usuario = str(input("¿Qué respuesta escoges? (Escribe la letra): "))

          #Comparamos respuesta del usuario y la respuesa correcta
          if respuesta_usuario == correcta:
            print("Respuesta correcta\n")
            suma_promedio += 1
            preg_correctas += 1
          else:
            print(f"Respuesta incorrecta, el resultado es {correcta}){correcta_sola}y tu pusiste la opción {respuesta_usuario})\n")
            preg_incorrectas += 1
          
          print(timer, end="\r")

        break

    #número de preguntas que existe en  la lista de preguntas, para luego tomar un promedio
    with open("preguntas.txt","r") as archivo_texto:
      line = archivo_texto.readlines()
      num_preguntas = len(line)

    promedio = suma_promedio / num_preguntas

    promedio = promedio /veces_func4

    #Creamos el archivo de texto para las calificaciones y respuestas correctas e incorrectas. Le asignamos una variable a cada una
    with open("califs.txt","w") as file:
      prom = file.write(str(round(promedio*100,2))+"\n") #Les pongo un caracter de escape para que esten renglon por renglon y no se junten
      preg_incorrectas = file.write(str(preg_incorrectas)+"\n")
      preg_correctas = file.write(str(preg_correctas)+"\n")
      
  #leer_archivo("califs.txt")
  veces_func4 += 1
  print(f"Esta función se ha llamado {veces_func4}")
  countdown(int(1200)) #Llamamos a la función del timer para que tome el tiempo límite que son 20 minutos

def reportar_calificaciones():
  global preguntas
  global promedio
  global suma_promedio 
  global preg_correctas
  global preg_incorrectas
  global veces_func5

  
  #Leemos el archivo de calificaciones y que lea los renglones. Como ya sabemos las posiciones, las concatenamos en un fstring para desplegar la tablita de resultados.
  with open("califs.txt","r") as f:
    pos = f.readlines()
    print(f"\nTotal de usuarios {veces_func4}\nPromedio de calificaciones {pos[0]}\nPreguntas correctas {pos[2]}\nPreguntas incorrectas {pos[1]}")

  veces_func5 += 1
  print(f"Esta función se ha llamado {veces_func5} veces")

def leer_archivo(nombre):
  with open(nombre, "r") as archivo_read:
    leer = archivo_read.readlines()

    if nombre == "ccalifs.txt":
      print("Este es el reporte de calificaciones")
      for num in range(len(leer)):
        if num == 0:
          print(f"{leer[num]} de calificación promedio")
        elif num == 1:
          print(f"{leer[num]} preguntas correctas")
        elif num == 2:
          print(f"{leer[num]} preguntas incorrectas")

    elif nombre == "preguntas.txt":
      print("Esta es la lista de preguntas con la nueva pregunta:")
      for num in range(len(leer)):
        print(f"Pregunta número {num+1} {leer[num]}")
    

def grabar_archivo(nombre, lista):
  with open(lista, "r") as archivo_a_leer:
    renglon = archivo_a_leer.readlines()

def salir():
  print("Muchas gracias por utilizar el programa, esperamos que te haya sido útil.")

main()
```
